```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Show all columns when printing dataframes
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 60)

print("Libraries loaded ✓")
```

```python
# Load main application file
app = pd.read_csv('../Data/application_train.csv')

# Load bureau data (past loans from other banks)
bureau = pd.read_csv('../Data/bureau.csv')

# Load previous applications at Home Credit
prev_app = pd.read_csv('../Data/previous_application.csv')

print("Files loaded ✓")
print(f"Application train : {app.shape}")
print(f"Bureau            : {bureau.shape}")
print(f"Previous apps     : {prev_app.shape}")
```

```python
#understand the target column 
# How many defaults vs non-defaults?
target_counts = app['TARGET'].value_counts()
target_pct = app['TARGET'].value_counts(normalize=True) * 100

print("Target distribution:")
print(f"  Repaid (0)   : {target_counts[0]:,} ({target_pct[0]:.1f}%)")
print(f"  Defaulted (1): {target_counts[1]:,} ({target_pct[1]:.1f}%)")

# Visualize it
plt.figure(figsize=(6, 4))
sns.countplot(x='TARGET', data=app, palette=['#2ecc71', '#e74c3c'])
plt.title('Loan Default Distribution')
plt.xticks([0, 1], ['Repaid', 'Defaulted'])
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Calculate missing value count and percentage for each column
missing = pd.DataFrame({
    'Missing Count': app.isnull().sum(),
    'Missing %': (app.isnull().sum() / len(app) * 100).round(2)
})

# Show only columns that have missing values, sorted by worst
missing = missing[missing['Missing Count'] > 0].sort_values('Missing %', ascending=False)

print(f"Columns with missing values: {len(missing)} out of {app.shape[1]}")
print("\nTop 20 columns with most missing data:")
print(missing.head(20))
```

```python
#seperate numerical and categorical columns

# Numerical columns
num_cols = app.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Categorical columns
cat_cols = app.select_dtypes(include=['object']).columns.tolist()

print(f"Numerical columns  : {len(num_cols)}")
print(f"Categorical columns: {len(cat_cols)}")
print(f"\nCategorical columns list:")
print(cat_cols)
```

```python
# These are the most important columns — look at them one by one

# Income distribution
print("Income stats:")
print(app['AMT_INCOME_TOTAL'].describe())

# Age (stored as negative days — we'll convert later)
print("\nAge in days (negative = past):")
print(app['DAYS_BIRTH'].describe())

# Convert to years to make sense of it
app['AGE_YEARS'] = abs(app['DAYS_BIRTH']) / 365
print("\nAge in years:")
print(app['AGE_YEARS'].describe())

# Employment length
print("\nEmployment days:")
print(app['DAYS_EMPLOYED'].describe())
# Note: you'll see a value of 365243 which is an anomaly — retired/unemployed people
```
